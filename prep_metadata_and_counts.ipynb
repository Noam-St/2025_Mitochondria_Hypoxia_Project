```python
# Fix peaks_counts by converting a [chr, start, end, id, counts] to [chr, start, end, counts] as required by DiffBind (https://bioconductor.org/packages/release/bioc/manuals/DiffBind/man/DiffBind.pdf#page=38&zoom=100,134,300)

from glob import glob
PATH = os.getcwd()
peaks_path = os.path.join(PATH, 'test_diffbind', 'peaks_count')
peak_files = glob(os.path.join(peaks_path, '*.txt'))
for peak_file in peak_files[1:]:
    peaks_cur = pd.read_csv(peak_file, sep = '\t', names = ['chr', 'start', 'end', 'peak_id', 'counts'], header = None, usecols = range(5))
    peaks_cur = peaks_cur.drop(columns = ['peak_id'])
    peaks_cur.to_csv(peak_file, sep = '\t', index = False, header = False)
```
